## **Load edge dataset using NetworkX**
Python에서 그래프(네트워크)를 그리는데 많이 사용되는 networkX를 사용하면 단순 edge list 정보로 쉽게 그래프 자료구조로 변환할 수 있다. 해당 예시에서는 Yeast PPI 네트워크의 edge list 파일로 networkX의 Graph 객체를 곧바로 생성한다

```python
from __future__ import division
from __future__ import print_function
import time
import numpy as np
import scipy.sparse as sp
import networkx as nx
from matplotlib import pylab
import matplotlib.pyplot as plt

#np.random.seed(0)

def load_data():
    G = nx.read_edgelist('yeast.edgelist')
    adj = nx.adjacency_matrix(G)
    return G, adj

print("Loading Yeast PPI network...")
G, adj = load_data()
```

```python
def load_yeast_seqs():
    return np.load("yeast.sequences.npz")['protein_seqs'][()]

print("Before: Graph Nodes (N): {}".format(len(G.nodes)))
print("Before: Graph Edges (E): {}".format(len(G.edges)))
print("Remove not exist nodes")
y2seqs = load_yeast_seqs()
for k, v in y2seqs.items():
    if v[0] == 0: 
        G.remove_node(k)

print("After : Graph Nodes (N): {}".format(len(G.nodes)))
print("After : Graph Edges (E): {}".format(len(G.edges)))
```

## **Open Source Python Lib - Graph Sampling**
RandomWork 알고리즘을 기반으로 하는 Graph sampling 공개 라이브러리를 활용. NetworkX의 Graph를 지원한다. 제공하는 sampling 방법은 아래와 같다
#### Sampling by Exploration
- Simple Random Walk Sampling (SRW)
- Random Walk Sampling with Fly Back Probability (RWF)
- Induced Subgraph Random Walk Sampling (ISRW)
- Snowball Sampling (SB)
- ForestFire Sampling (FF) 
- Metropolis Hastings Random Walk Sampling (MHRW)
- Induced Metropolis Hastings Random Walk Sampling (Induced-MHRW)

#### Edge Sampling
- Total Induction Edge Sampling (TIES)

### **Updated ForestFire Sampling (FF)**

```python
#from Graph_Sampling import ForestFire
# https://github.com/Ashish7129/Graph_Sampling/blob/master/Graph_Sampling/ForestFire.py
from copy import deepcopy
import random
import time
from collections import defaultdict

# https://github.com/Ashish7129/Graph_Sampling/blob/master/Graph_Sampling/ForestFire.py
class ForestFire():
    def __init__(self):
        self.G1 = nx.Graph()
        self.emptyG = nx.Graph()

    def forestfire(self,G,size):
        list_nodes=list(G.nodes())
        #print(len(G))
        dictt = set()
        visits = defaultdict(lambda: False)
        random_node = random.sample(set(list_nodes),1)[0]
        #print(random_node)
        q = set() #q = set contains the distinct values
        q.add(random_node)
        while(len(self.G1.nodes())<size):
            if(len(q)>0):
                initial_node = q.pop()
                if(initial_node not in dictt):
                    dictt.add(initial_node)
                    neighbours = list(G.neighbors(initial_node))
                    #print(list(G.neighbors(initial_node)))
                    np = random.randint(1,len(neighbours))
                    #print(np)
                    #print(neighbours[:np])
                    for x in neighbours[:np]:
                        if(len(self.G1.nodes())<size):
                            self.G1.add_edge(initial_node,x)
                            q.add(x)
                        else:
                            break
                else:
                    continue
            else:
                random_node = random.sample(set(list_nodes) and dictt,1)[0]
                if visits[random_node] == True:
                    print("redundant node {}".format(random_node))
                    return self.emptyG
                else:
                    q.add(random_node)
                    visits[random_node] = True
                    print("add {}".format(random_node))
                
        q.clear()
        return self.G1
    
# input - graph-sampling class, target Graph, number of nodes, rows, cols
def visualize_ForestFire(target_graph, nodes_to_sample=10, sample_rows=10, sample_cols=10):
    # DO in the function
    
    
    num_samples = sample_rows * sample_cols
    
    node_color = '#000080'
    node_size = 50
    node_alpha = .6

    edge_color = '#483D8B'
    edge_alpha = 0.3
    edge_style = 'dashed'
    edge_width = 2
    
    repeat_number = 10
    
    for i in range(repeat_number):
        for r in range(sample_rows):
            plt.figure(figsize=(30,6))
            for c in range(sample_cols):
                sample_idx = r*sample_cols + c + 1
                plt.subplot(sample_rows,sample_cols, sample_idx)
                gsampler = ForestFire()
                sampled_graph = gsampler.forestfire(target_graph, nodes_to_sample)

                pos = nx.spring_layout(sampled_graph)
                nx.draw_networkx_nodes(sampled_graph, pos, node_size=node_size, alpha=node_alpha, node_color=node_color)
                nx.draw_networkx_edges(sampled_graph, pos, width=edge_width, alpha=edge_alpha, edge_color=edge_color, style=edge_style)
                plt.axis('on')
                plt.title('ForestFire Sampling Sample {}'.format(sample_idx))
                #print("loading... {}/{} / nodes : {}".format(r*sample_cols + c + 1, num_samples, sampled_graph.nodes))
                #print("loading... {}/{} ".format(r*sample_cols + c + 1, num_samples))

            plt.savefig("imgs/yeast-ppi-rnd-samples-{}.png".format(i))
            plt.show()

            

visualize_ForestFire(target_graph = G, nodes_to_sample=100, sample_rows=1, sample_cols=5)
```
